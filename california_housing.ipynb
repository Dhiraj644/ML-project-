```python
from sklearn.datasets import fetch_california_housing
california=fetch_california_housing(as_frame=True)
california_df=california['frame']
california_df
```

```python
california_df.to_csv("california_df.csv")
```

```python
x=california_df.drop('MedHouseVal',axis=1)
y=california_df['MedHouseVal']
```

```python
california_df.info
```

```python
california_df.shape
```

```python
california_df.isnull().sum()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=[]
for i in range (len(x.columns)):
    vif_data.append(variance_inflation_factor(x.values,i))
vif_data
```

```python
import pandas as pd
vif_df=pd.DataFrame(vif_data,columns=["VIF"],index=x.columns)
```

```python
vif_df
```

```python
con1=vif_df[vif_df['VIF']<15].index
con1=con1.to_list()
```

```python
con1
```

```python
con2=vif_df[vif_df['VIF']>15].index
con2=con2.to_list()
con2
```

```python
con2=["AveRooms","Latitude"]
```

```python
final_con=con1+con2
final_con
```

```python
import warnings
warnings.filterwarnings('ignore')
final_df=x[final_con]
final_df['MedHouseVal']=y
final_df
```

```python
x=final_df.drop('MedHouseVal',axis=1)
y=final_df['MedHouseVal']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scale=ss.fit_transform(x)
x_scale
```

```python
x_scale_df=pd.DataFrame(x_scale,columns=x.columns)
```

```python
x_scale_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scale_df,y,test_size=0.2,random_state=12)
```

```python
from sklearn.linear_model import LinearRegression

lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error

print(r2_score(y_test,y_pred))
```
